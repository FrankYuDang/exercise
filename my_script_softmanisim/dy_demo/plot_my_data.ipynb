我们现在来分析得到的数据，利用 pandas, pathlib, matplotlib, numpy, scipy 进行分析

```python
import pandas as pd
import numpy as np
from pathlib import Path 
import matplotlib as mpl
import matplotlib.pyplot as plt
# %matplotlib inline 

mpl.rcParams['font.family'] = 'SimSun'
mpl.rcParams['font.size'] = 10
```

## load the result data
This is used to draw the typical setup.

```python
# load data from .\output__data\my_yomo
filename = 'yomo_run_log_20250428_223128.csv'
# data_path = Path('C:/Document/Projects/SoftManiSim/output_data/my_yomo')
data_path = Path('../../output_data/my_yomo_refactored')
# print(data_path.resolve()) 打印路径全名
# data_path = Path('.\output__data\my_yomo')
df = pd.read_csv(data_path / filename)
print(df.head())
num_plots = 3  # 减去第一列的时间戳
```

## Plot the results of cable length

```python
fig, axes = plt.subplots(nrows=1, ncols = num_plots, figsize=(2.5*num_plots, 3))
for i in range(num_plots):
    ax = axes[i]
    data = df.iloc[:, i+1]  # Skip the first column (time)
    time = df.iloc[:, 0]  # First column (time)
    
    # 在当前的ax上绘图
    ax.plot(time, data, label=df.columns[i+1], color='blue')
    ax.set_title(df.columns[i+1])
    ax.set_xlabel('Time (s)')
    ax.set_ylabel(df.columns[i+1])
    ax.grid(True, linestyle='--', alpha=0.5)

# 调整布局并显示
fig.suptitle('Yomo Simulation Results', fontsize=16)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the title
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=num_plots, figsize=(2.5*num_plots, 3))
for i in range(num_plots):
    ax = axes[i]
    data = df.iloc[:, i+4]
    time = df.iloc[:, 0]

    ax.plot(time, data, label=df.columns[i+4], color='red')
    ax.set_title(df.columns[i+4])
    ax.set_xlabel('Time (s)')
    ax.set_ylabel(df.columns[i+4])
    ax.grid(True,linestyle='--', alpha=0.4)

fig.suptitle('Yomo Simulation results - Head Position')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
    
```

```python
plt.figure(1)
plt.clf
plt.plot(df['head_y'], df['head_z'],'r*', markersize=5)
plt.title('末端点的轨迹')
plt.xlabel('y 方向')
plt.ylabel('x 方向')
plt.show()
```
