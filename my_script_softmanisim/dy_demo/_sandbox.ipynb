```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'SimSun'
mpl.rcParams['font.size'] = 16

t = 0.0
dt = 0.01
# T = range(t, 10.0, dt) range() needs integer arguments
T = np.arange(t, 10.0, dt)
yomo_cable_2 = .005*np.sin(0.5*np.pi*T+2+5)
yomo_cable_3 = .005*np.cos(0.5*np.pi*T+2+5)
```

```python
plt.figure(figsize=(10, 5), dpi=80)
plt.plot(T, yomo_cable_2, label='yomo_cable_2', color='blue', linewidth=1.0, linestyle='-')
plt.plot(T, yomo_cable_3, 'r*' , markersize=5, label='yomo_cable_3', )
plt.title('yomo_cable_2')
plt.xlabel('Time (s)')
plt.ylabel('Action[2]')
plt.legend(loc='best')
plt.show()
```
